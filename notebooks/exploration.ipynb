# Soybean Price Exploration
Notebook inicial para explorar o dataset de preços da soja.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/soybean_prices.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['date'], df['price'])
plt.title('Preço da soja ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show()
```
